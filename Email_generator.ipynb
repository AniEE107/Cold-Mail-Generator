```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

llm = ChatGroq(
    model="llama3-70b-8192",  # ✅ correct name
    temperature=0.7,
    max_tokens=1024,
    max_retries=2
)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert technical assistant."),
    ("user", "{input}")
])

chain = prompt | llm

response = chain.invoke({"input": "The First Animal to landn was ..."})
print(response.content)
```

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Fetch page HTML directly
url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url)
html = response.text

# Step 2: Parse with BeautifulSoup
soup = BeautifulSoup(html, "html.parser")

# Step 3: Find all job cards
jobs = soup.find_all("div", class_="card-content")
print(f"✅ Number of job cards found: {len(jobs)}\n")

# Step 4: Loop through and print job details
for job in jobs:
    title = job.find("h2", class_="title").text.strip()
    company = job.find("h3", class_="company").text.strip()
    location = job.find("p", class_="location").text.strip()
    date = job.find("time")["datetime"]
    
    print(f"🧾 Title: {title}")
    print(f"🏢 Company: {company}")
    print(f"📍 Location: {location}")
    print(f"📅 Date: {date}")
    print("-" * 50)
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
import json

template = """
### WEBPAGE CONTENT:
{page_data}

### INSTRUCTIONS:
Extract all job listings with these fields:
1. "role": Job title (from <h2> elements)
2. "company": Company name (from <h3> elements)
3. "location": Job location (from <p class="location">)
4. "date": Post date (from <time> elements)
5. "description": Job description (from div with class="content")

For any missing fields, use:
- "" for strings
- [] for lists
- null for completely missing data

Return ONLY valid JSON.
"""

prompt = PromptTemplate(
    template=template,
    input_variables=["page_data"]
)

# Set up the extraction chain
chain = prompt | llm | JsonOutputParser()

try:
    # Process the page data
    jobs = chain.invoke({"page_data": page_data})
    
    # Clean empty fields
    for job in jobs:
        job.setdefault("experience", "")
        job.setdefault("skills", [])
        job.setdefault("description", "")
    
    print(json.dumps(jobs, indent=2))
    
except Exception as e:
    print(f"Error: {e}")
    print("Sample of problematic data:")
    print(page_data[:1000])
```
